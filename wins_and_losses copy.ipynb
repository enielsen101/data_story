```python
## slice off the wins and losses

df_outcomes = df_all[["Year", "Cert of Rep (Win)", "Cert of Results (Loss)"]].copy()
df_outcomes
```

```python
## create a new column based on win and loss

df_outcomes['win_or_loss'] = df_outcomes.apply(
    lambda row: "Win" if row["Cert of Rep (Win)"] == 'WON' else ("Loss" if row["Cert of Results (Loss)"] == 'LOSS' else None),
    axis=1
)
df_outcomes
```

```python
summary = df_outcomes.groupby(['Year', 'win_or_loss']).size().reset_index(name='count')
summary
```

```python
pivot = summary.pivot(index='Year', columns='win_or_loss', values='count').fillna(0).astype(int)
print(pivot)
```

```python
pivot['Rate of Win'] = pivot['Win'] / (pivot['Win'] + pivot['Loss']) * 100
pivot['Rate of Win'] = pivot['Rate of Win'].round(1)
pivot
```

```python
pivot.to_csv("wins_and_losses.csv")
```
