```python
# Filter rows by RC to get only union filing petitions
filtered_df = df[df["Case Number"].str.contains("-RC-")].copy()
```

```python
## Filter df by date filed
filtered_df['Date Filed'] = pd.to_datetime(filtered_df['Date Filed'], errors='coerce')
```

```python
filtered_df
```

```python
## Filter df by FY 2021 - 2024

date_filed_df = filtered_df.query("(@pd.to_datetime('2020-10-01') <= `Date Filed` <= @pd.to_datetime('2021-09-30')) or \
     (@pd.to_datetime('2021-10-01') <= `Date Filed` <= @pd.to_datetime('2022-09-30')) or \
     (@pd.to_datetime('2022-10-01') <= `Date Filed` <= @pd.to_datetime('2023-09-30')) or \
     (@pd.to_datetime('2023-10-01') <= `Date Filed` <= @pd.to_datetime('2024-09-30'))"
)
```

```python
## Get info on new dataset - #s do not match NLRB reports, so something went wrong
date_filed_df.info()
```

```python
## Put 2021 - 2024 into dataframes
## First try: df_2021 = date_filed_df[date_filed_df["Date Filed"].dt.date (2022-09-30) <= 'Date Filed' <= (2021-10-01)]
## Second try

df_2021 = date_filed_df[
    (date_filed_df["Date Filed"] >= "2020-10-01") & (date_filed_df["Date Filed"] <= "2021-09-30")].copy()
df_2022 = date_filed_df[
    (date_filed_df["Date Filed"] >= "2021-10-01") & (date_filed_df["Date Filed"] <= "2022-09-30")].copy()
df_2023 = date_filed_df[
    (date_filed_df["Date Filed"] >= "2022-10-01") & (date_filed_df["Date Filed"] <= "2023-09-30")].copy()
df_2024 = date_filed_df[
    (date_filed_df["Date Filed"] >= "2023-10-01") & (date_filed_df["Date Filed"] <= "2024-09-30")].copy()
```

```python
## realized I need a simpler table with just the number of filings per year

## df_2020s.groupby("Case Number")["Date Filed"].sum().to_frame("Filings")
## didn't work because can't group by dates
```

```python
## instead created a new column in the individual year dfs

df_2021["Fiscal Year"] = 2021
df_2022["Fiscal Year"] = 2022
df_2023["Fiscal Year"] = 2023
df_2024["Fiscal Year"] = 2024
```

```python
## see if it worked
df_2021
```

```python
## create new df with 2021-2024 and call it

all_years_df = pd.concat([df_2021, df_2022, df_2023, df_2024], ignore_index=True)
all_years_df
```

```python
## create new df that groups by Fiscal Year column
## df.groupby("School Type")["dl_sub_$"].sum().to_frame("total $ sub DL")
all_years_df.groupby("Fiscal Year")["Fiscal Year"].count().to_frame("NLRB Filings")
```
