```python
import pandas as pd
```

```python
## import the csv

df = pd.read_csv("data_raw/NLRB_all_cases.csv", low_memory=False)
```

```python
## get info

df.info()
```

```python
## call head

df.head()
```

```python
## create df for companies

participants = df['Participants']
participants
```

```python
## sort for Trader Joe's, call it

TJ = df[df['Participants'].str.contains('Trader Joe', na=False)]
TJ
```

```python
## change display to read Participants column

pd.set_option('display.max_colwidth', None)
TJ
```

```python
## sort by date

TJ['Date Filed'] = pd.to_datetime(df['Date Filed'], errors='coerce')
```

```python
## With help from AI, made a copy (because it's a slice):

TJ = TJ.copy()
```

```python
## Filter by date filed with date format:

TJ['Date Filed'] = pd.to_datetime(TJ['Date Filed'], format='%m/%d/%y', errors='coerce')
```

```python
## Print info:

TJ['Year Filed'] = TJ['Date Filed'].dt.year
cases_per_year = TJ['Year Filed'].value_counts().sort_index()
print(cases_per_year)
```

```python
## export

cases_per_year.to_excel('cases_per_year.xlsx', index=False)
```

```python
# Convert series to df 

summary = cases_per_year.reset_index()
summary.columns = ['Year', 'complaints filed']
```
